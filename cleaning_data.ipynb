```python
import pandas as pd
import numpy as np
```

```python
# Charger toutes les données d'une feuille spécifique
df = pd.read_excel("data/Table Ciqual 2020_FR_2020 07 07.xls", sheet_name="compo")
```

```python
df.columns.tolist()
```

```python
# Liste des colonnes à supprimer
colonnes_a_supprimer = ['alim_grp_code',
 'alim_ssgrp_code',
 'alim_ssssgrp_code',
 'alim_code',
 'alim_nom_sci',
 'Energie, Règlement UE N° 1169/2011 (kJ/100 g)',
 'Energie, N x facteur Jones, avec fibres  (kJ/100 g)',
 'Energie, Règlement UE N° 1169/2011 (kcal/100 g)',
 'Protéines, N x 6.25 (g/100 g)',]
```

```python
# Supprimer les colonnes
df_suppr = df.drop(columns=colonnes_a_supprimer)

print("NB colonnes avant: {}".format(df.shape[1]))
print("NB colonnes après: {}".format(df_suppr.shape[1]))
```

```python
dict_rename = {'alim_grp_nom_fr': 'Groupe nom', 
               'alim_ssgrp_nom_fr': 'Sous-groupe nom', 
               'alim_ssssgrp_nom_fr':'Sous-sous-groupe nom',
               'alim_nom_fr':'Nom',
               'Energie, N x facteur Jones, avec fibres  (kcal/100 g)':'Calories (kcal/100 g)',
               'Protéines, N x facteur de Jones (g/100 g)':'Protéines (g/100 g)'}
# Dictionnaire de renaming : {ancien_nom: nouveau_nom}
df_suppr.rename(columns=dict_rename, inplace=True)
```

```python
df_suppr.columns.tolist()
```

```python
df_suppr.iloc[:, 4:]
```

```python
df_suppr.iloc[:, 4].unique()
```

```python
# objectifs:
# - remplacer valeurs '-' par null
# - remplacer valeurs 'traces' par une valeur très faible (0.01)
# - remplacer '< ' par une chaîne vide 

for i in range(4,67):
    df_suppr.iloc[:, i] = df_suppr.iloc[:, i].replace({'-': np.nan, 'traces': 0.01})

    df_suppr.iloc[:, i] = df_suppr.iloc[:, i].str.replace('< ', '')
    df_suppr.iloc[:, i] = df_suppr.iloc[:, i].str.replace(',', '.')

    #df_suppr.iloc[:, i] = df_suppr.iloc[:, i].fillna(0)
    
    df_suppr.iloc[:, i] = df_suppr.iloc[:, i].astype(float)
```

```python
df_suppr.iloc[:, 4:]
```

```python
df_suppr.iloc[0,0] = "desserts"
df_suppr.iloc[0,1] = "-"
df_suppr.iloc[0,2] = "-"
df_suppr.head(5)
```

```python
# Exporter en fichier Excel (format .xlsx)
df_suppr.to_excel('data/Table Ciqual nettoyer.xlsx', sheet_name='Composition nutritionnelle', index=True)

print("Nettoyage fini...")
```
